```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from tensorflow import keras
```

```python
from keras.datasets import boston_housing
```

```python
(X_train,y_train),(X_test,y_test)=boston_housing.load_data() #for inbuilt datasets
```

```python
#normalize the data for Model
X_train = tf.keras.utils.normalize(X_train,axis=1)
X_test = tf.keras.utils.normalize(X_test,axis=1)
```

```python
n_cols =X_train.shape[1]  # number of columns
```

```python
# The input shape specifies the number of rows and columns in the input. The number of columns in our input is stored in ‘n_cols’. 
#There is nothing after the comma which indicates that there can be any amount of rows.
```

```python
#model -sequential
from keras.models import Sequential
model=Sequential()
#adding layers for the model
from keras.layers import Dense
model.add(Dense(64,activation="relu",input_shape=(13,)))
model.add(Dense(64,activation="relu"))
model.add(Dense(64,activation="relu"))
model.add(Dense(1))
model.summary()
```

```python
#training
import keras.metrics 
from keras.optimizers import rmsprop
opt=rmsprop(lr=0.001)
model.compile(optimizer=opt,loss="mean_squared_error",metrics=(["mse"])) # mse- mean sqaure error
```

```python
history=model.fit(X_train,y_train,validation_split=0.2,batch_size=30, epochs=1000)
```

```python
plt.plot(history.history["loss"])
```
